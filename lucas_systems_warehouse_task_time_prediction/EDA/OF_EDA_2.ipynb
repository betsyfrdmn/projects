```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Plot distribution to find the "gap" for break thresholds
plt.figure(figsize=(10, 6))
# Filter out 0s and extreme outliers for a better view
sns.histplot(of_df[of_df['Pick_Time_Hours'] < 1]['Pick_Time_Hours'], bins=100, kde=True)
plt.axvline(x=of_df['Pick_Time_Hours'].quantile(0.95), color='r', linestyle='--', label='95th Percentile')
plt.title("Distribution of Pick Times (Under 1 Hour)")
plt.xlabel("Hours")
plt.show()

# Based on EDA, you can then 'flag' or 'reset' time delta
# Example: If a gap > 0.5 hours, it's a break, not travel.
threshold = 0.5 
of_df['Is_Break'] = of_df['Pick_Time_Hours'] > threshold
```

```python
# 1. Extract Level (assuming format Aisle|Bay|Level|...)
of_df['Level'] = of_df['Location'].str.split('|').str[2]

# 2. Identify consecutive picks at the same location (Stay-Picks)
of_df['Is_Stay_Pick'] = of_df['Location'] == of_df.groupby(['UserID', 'Shift'])['Location'].shift(1)

# 3. Aggregate consecutive picks of the same product
# This creates a 'Group ID' that changes every time the ProductCode changes
of_df['Product_Group'] = (of_df['ProductCode'] != of_df.groupby('UserID')['ProductCode'].shift()).cumsum()

# Compare the mean Pick_Time_Secs for levels
level_eda = of_df.groupby('Level')['Pick_Time_Secs'].mean().sort_values()
print("Average Pick Time by Level:\n", level_eda)
```

```python
# Aggregate by User to find the baseline
user_performance = of_df.groupby('UserID').agg({
    'Pick_Time_Secs': ['mean', 'std', 'median'],
    'Distance': 'sum',
    'ActivityCode': 'count'
})

# Calculate individual velocity (Distance / Time)
user_performance['Velocity'] = user_performance[('Distance', 'sum')] / (user_performance[('Pick_Time_Secs', 'mean')] * user_performance[('ActivityCode', 'count')])

# Display distribution of velocities to find the "Average User"
sns.boxplot(x=user_performance['Velocity'])
plt.title("Distribution of User Velocities")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = of_activity.copy()
# 1. Prepare the data
# Assuming df_activity is your first table
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['Day'] = df['Timestamp'].dt.date

# 2. Daily Aggregation
daily_stats = df.groupby('Day').agg(
    picktasks=('ActivityCode', 'count'),          # Total rows
    assignments=('AssignmentID', 'nunique'),     # Distinct AssignmentIDs
    total_quantity=('Quantity', 'sum'),          # Sum of Quantity
    workers=('UserID', 'nunique')                # Distinct UserIDs
).reset_index()

# 3. Create the Visualization
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
plt.subplots_adjust(hspace=0.4, wspace=0.3)

metrics = [
    ('picktasks', 'Picktasks (rows)', axes[0, 0]),
    ('assignments', 'Distinct AssignmentID', axes[0, 1]),
    ('total_quantity', 'Total Quantity', axes[1, 0]),
    ('workers', 'Distinct Workers (UserID)', axes[1, 1])
]

for col, title, ax in metrics:
    ax.plot(daily_stats['Day'], daily_stats[col])
    ax.set_title(title, fontsize=10)
    ax.set_ylabel('Count' if 'quantity' not in col else 'Sum')
    ax.set_xlabel('Day')
    # Rotate x-labels if they get crowded
    plt.setp(ax.get_xticklabels(), rotation=0)

plt.tight_layout()
plt.show()

# 4. Generate the Summary Table
summary_table = daily_stats[['picktasks', 'assignments', 'total_quantity', 'workers']].agg(['mean', 'min', 'max']).T
print(summary_table)
```

```python
# sort by AssignmentID then Timestamp and compute difference to previous timestamp within same assignment
df = df.sort_values(["AssignmentID", "Timestamp"])
df["time_diff"] = df.groupby("AssignmentID")["Timestamp"].diff()
df["time_diff_sec"] = df["time_diff"].dt.total_seconds()

df.head()
```

```python
# aggregate time_diff_sec by AssignmentID, LocationID, WorkCode
agg = (
    df.dropna(subset=["time_diff_sec"])
      .groupby(["AssignmentID", "LocationID", "WorkCode"], as_index=False)
      .agg(
          mean_time=("time_diff_sec", "mean"),
          sum_time=("time_diff_sec", "sum"),
          median_time=("time_diff_sec", "median"),
          n=("time_diff_sec", "count"),
      )
)

# overlayed KDEs per WorkCode (using mean_time)
metric = "mean_time"

plt.figure(figsize=(10, 6))
workcodes = sorted(agg["WorkCode"].unique())
for wc in workcodes:
    vals = agg.loc[(agg["WorkCode"] == wc) & (agg["n"] > 1), metric]
    if len(vals) > 1:
        sns.kdeplot(vals, label=f"WorkCode {wc}", bw_adjust=1, common_norm=False, alpha=0.8)

plt.xlabel(metric)
plt.ylabel("Density")
plt.title(f"Overlayed KDE of {metric} by WorkCode")
plt.legend(title="WorkCode")
plt.tight_layout()
plt.show()
```

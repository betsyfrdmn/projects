```python
import pandas as pd
import re

# --- Keywords for Filtering ---
KEYWORDS = ['islam', 'muslim', 'muslims', 'islamic',
            'christianity', 'christian', 'christians',
            'judaism', 'jew', 'jews', 'jewish']

# --- Filtering Function ---
def filter_by_keywords(df, column, keywords):
    """
    Filters a DataFrame to keep rows where the specified column contains
    at least one of the keywords (case-insensitive).
    """
    # 1. Create a regex pattern combining all keywords with OR condition (|)
    #    're.escape' handles any special characters in the keywords.
    escaped_keywords = [re.escape(k) for k in keywords]
    pattern = '|'.join(escaped_keywords)
    
    # 2. Apply the filter using str.contains() with case=False (case-insensitive)
    #    na=False ensures missing values in bodyContent don't cause errors.
    mask = df[column].str.contains(pattern, case=False, na=False)
    
    # 3. Apply the mask to the DataFrame
    df_filtered = df[mask].copy()
    
    return df_filtered

# --- Execution ---
df_filtered = filter_by_keywords(df, 'bodyContent', KEYWORDS)
df_filtered.to_csv("filtered_articles.csv", index=False)
```
